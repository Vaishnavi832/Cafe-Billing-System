```python
class Task:
    def __init__(self, id, title, description, status, due_date):
        self.id = id
        self.title = title
        self.description = description
        self.status = status
        self.due_date = due_date

    def __str__(self):
        return f"ID: {self.id} | Title: {self.title} | Status: {self.status} | Due Date: {self.due_date}"

class TaskManager:
    def __init__(self):
        self.tasks = {}
        self.next_id = 1

    def create_task(self, title, description, status, due_date):
        task = Task(self.next_id, title, description, status, due_date)
        self.tasks[self.next_id] = task
        self.next_id += 1
        return f"Task {task.id} created successfully!"

    def read_all_tasks(self):
        if not self.tasks:
            print("No tasks available")
        for task in self.tasks.values():
            print(task)

    def read_task(self, task_id):
        task = self.tasks.get(task_id)
        if task:
            return task
        else:
            return f"Task with ID {task_id} not found."

    def update_task(self, task_id, title=None, description=None, status=None, due_date=None):
        task = self.tasks.get(task_id)
        if task:
            if title:
                task.title = title
            if description:
                task.description = description
            if status:
                task.status = status
            if due_date:
                task.due_date = due_date
            return f"Task {task_id} updated successfully!"
        else:
            return f"Task with ID {task_id} not found."

    def delete_task(self, task_id):
        if task_id in self.tasks:
            del self.tasks[task_id]
            return f"Task {task_id} deleted successfully"
        else:
            return f"Task with ID {task_id} not found"

# Testing
manager = TaskManager()
print(manager.create_task("Python", "This book contains all basics", "pending", "02-05-2025"))
print(manager.read_task(1))
print(manager.delete_task(1))
```

```python
class Order:
    def __init__(self, id, customer_name, items, status, order_time):
        self.id = id
        self.customer_name = customer_name
        self.items = items
        self.status = status
        self.order_time = order_time

    def __str__(self):
        item_list = ", ".join(self.items)
        return f"Order ID: {self.id} | Customer: {self.customer_name} | Items: [{item_list}] | Status: {self.status} | Time: {self.order_time}"

class CafeOrderManager:
    def __init__(self):
        self.orders = {}
        self.next_id = 1

    def create_order(self, customer_name, items, status, order_time):
        order = Order(self.next_id, customer_name, items, status, order_time)
        self.orders[self.next_id] = order
        self.next_id += 1
        return f"Order {order.id} created successfully!"

    def read_all_orders(self):
        if not self.orders:
            print("No orders available.")
        else:
            for order in self.orders.values():
                print(order)

    def read_order(self, order_id):
        order = self.orders.get(order_id)
        if order:
            return order
        else:
            return f"Order with ID {order_id} not found."

    def update_order(self, order_id, customer_name=None, items=None, status=None, order_time=None):
        order = self.orders.get(order_id)
        if order:
            if customer_name:
                order.customer_name = customer_name
            if items:
                order.items = items
            if status:
                order.status = status
            if order_time:
                order.order_time = order_time
            return f"Order {order_id} updated successfully!"
        else:
            return f"Order with ID {order_id} not found."

    def delete_order(self, order_id):
        if order_id in self.orders:
            del self.orders[order_id]
            return f"Order {order_id} deleted successfully."
        else:
            return f"Order with ID {order_id} not found."


cafe = CafeOrderManager()


print(cafe.create_order("Alice", ["Cappuccino", "Bagel"], "Pending", "10:15 AM"))
print(cafe.create_order("Bob", ["Espresso", "Croissant"], "Preparing", "10:30 AM"))
print(cafe.create_order("Charlie", ["Latte", "Muffin"], "Pending", "10:45 AM"))


print("\nAll Orders:")
cafe.read_all_orders()


print("\nReading Order ID 2:")
print(cafe.read_order(2))

print("\nUpdating Order ID 2:")
print(cafe.update_order(2, status="Ready", items=["Espresso", "Croissant", "Orange Juice"]))


print("\nDeleting Order ID 1:")
print(cafe.delete_order(1))


print("\nAll Orders After Update & Delete:")
cafe.read_all_orders()
```
