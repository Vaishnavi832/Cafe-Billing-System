```python
import pandas as pd
import numpy as np

f1 = pd.read_csv("D:\\python_itp\\loan_sanction_test (3).csv")


print("ðŸ”¹ Dataset Preview:")
print(f1.head())  # shows first 5 rows

print("\nðŸ”¹ Basic Info:")
print(f1.info())  # column info (types, non-null counts)

print("\nðŸ”¹ Statistical Summary:")
print(f1.describe())  # summary of numeric columns

# Check for missing values
print("\nðŸ”¹ Missing Values in Each Column:")
print(f1.isnull().sum())

#  Drop missing values (optional)
f1_clean = f1.dropna()
print("\nðŸ”¹ After Dropping Missing Values:")
print(f1_clean.head())

#  Select specific columns
if 'LoanAmount' in f1.columns and 'ApplicantIncome' in f1.columns:
    print("\nðŸ”¹ Loan Amount and Applicant Income Columns:")
    print(f1[['LoanAmount', 'ApplicantIncome']].head())

#  Filter data based on a condition
if 'Gender' in f1.columns:
    males = f1[f1['Gender'] == 'Male']
    print(f"\nðŸ”¹ Total Male Applicants: {len(males)}")
    print(males.head())

#  Grouping and Aggregation
if 'Education' in f1.columns and 'LoanAmount' in f1.columns:
    print("\nðŸ”¹ Average Loan Amount by Education Level:")
    print(f1.groupby('Education')['LoanAmount'].mean())

# Add a new calculated column (e.g., EMI or ratio)
if 'ApplicantIncome' in f1.columns and 'LoanAmount' in f1.columns:
    f1['Income_to_Loan_Ratio'] = f1['ApplicantIncome'] / f1['LoanAmount']
    print("\nðŸ”¹ Data with New 'Income_to_Loan_Ratio' Column:")
    print(f1[['ApplicantIncome', 'LoanAmount', 'Income_to_Loan_Ratio']].head())

#  Sort data
if 'LoanAmount' in f1.columns:
    sorted_f1 = f1.sort_values(by='LoanAmount', ascending=False)
    print("\nðŸ”¹ Top 5 Highest Loan Amounts:")
    print(sorted_f1.head())

#  Save cleaned/modified data to a new file
f1.to_csv("D:\\python_itp\\loan_sanction_cleaned.csv", index=False)
print("\nâœ… Cleaned file saved as 'loan_sanction_cleaned.csv'")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read CSV file
f1 = pd.read_csv("D:\\python_itp\\loan_sanction_test (3).csv")
```

```python
# Display first few rows
print(f1.head())
```

```python
# Line plot
plt.plot(f1['ApplicantIncome'])
plt.title("Applicant Income Trend")
plt.xlabel("Index")
plt.ylabel("Income")
plt.show()
```

```python
#  Histogram
if 'LoanAmount' in f1.columns:
    plt.hist(f1['LoanAmount'], bins=10, edgecolor='black')
    plt.title("Loan Amount Distribution")
    plt.xlabel("Loan Amount")
    plt.ylabel("Frequency")
    plt.show()
```

```python
if 'Gender' in f1.columns:
    gender_counts = f1['Gender'].value_counts()
    plt.bar(gender_counts.index, gender_counts.values, color='teal')
    plt.title("Count of Applicants by Gender")
    plt.xlabel("Gender")
    plt.ylabel("Count")
    plt.show()   #this is for bar chart
```

```python
#  Pie chart
if 'Education' in f1.columns:
    edu_counts = f1['Education'].value_counts()
    plt.pie(edu_counts, labels=edu_counts.index, autopct='%1.1f%%')
    plt.title("Education Distribution")
    plt.show()
```

```python
#  Box plot (outlier detection)
if 'LoanAmount' in f1.columns:
    plt.boxplot(f1['LoanAmount'].dropna())
    plt.title("Loan Amount Outliers")
    plt.show()
```

```python
#  Scatter plot
if 'ApplicantIncome' in f1.columns and 'LoanAmount' in f1.columns:
    plt.scatter(f1['ApplicantIncome'], f1['LoanAmount'])
    plt.title("Applicant Income vs Loan Amount")
    plt.xlabel("Applicant Income")
    plt.ylabel("Loan Amount")
    plt.show()
```

```python
#  Horizontal bar chart
if 'Property_Area' in f1.columns:
    area_counts = f1['Property_Area'].value_counts()
    plt.barh(area_counts.index, area_counts.values)
    plt.title("Loan Count by Property Area")
    plt.xlabel("Count")
    plt.ylabel("Property Area")
    plt.show()
```

```python

#  Multiple line plot
if 'ApplicantIncome' in f1.columns and 'CoapplicantIncome' in f1.columns:
    plt.plot(f1['ApplicantIncome'], label='Applicant Income')
    plt.plot(f1['CoapplicantIncome'], label='Coapplicant Income')
    plt.title("Applicant vs Coapplicant Income")
    plt.legend()
    plt.show()
```

```python
#  Subplots
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
if 'LoanAmount' in f1.columns:
    ax[0].hist(f1['LoanAmount'], bins=10)
    ax[0].set_title("Loan Amount Distribution")
if 'ApplicantIncome' in f1.columns:
    ax[1].hist(f1['ApplicantIncome'], bins=10)
    ax[1].set_title("Applicant Income Distribution")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
```

```python

f1 = pd.read_csv("D:\\python_itp\\loan_sanction_test (3).csv")
```

```python
applicant_income = f1['ApplicantIncome'].dropna().to_numpy()
coapplicant_income = f1['CoapplicantIncome'].dropna().to_numpy()
loan_amount = f1['LoanAmount'].dropna().to_numpy()
loan_term = f1['Loan_Amount_Term'].dropna().to_numpy()
credit_history = f1['Credit_History'].dropna().to_numpy()
```

```python
print("Average Applicant Income:", np.mean(applicant_income))
print("Median Coapplicant Income:", np.median(coapplicant_income))
print("Mean Loan Amount:", np.mean(loan_amount))
print("Loan Amount Standard Deviation:", np.std(loan_amount))
print("Median Loan Term (months):", np.median(loan_term))
print("Average Credit History:", np.mean(credit_history))
```

```python
print("Median Coapplicant Income:", np.median(coapplicant_income))
```

```python
print("Variance LoanAmount:", np.var(loan_amount))
print("Max LoanAmount:", np.max(loan_amount))
print("Min LoanAmount:", np.min(loan_amount))
print("Sum of LoanAmount:", np.sum(loan_amount))
print("90th Percentile of LoanAmount:", np.percentile(loan_amount, 90))
```

```python
loan_range = np.ptp(loan_amount)
total_loan = np.sum(loan_amount)
print("\nLoan Amount Range:", loan_range)
print("Total Loan Amount:", total_loan)
```

```python
sorted_loans = np.sort(loan_amount)
max_index = np.argmax(loan_amount)
min_index = np.argmin(loan_amount)
print("Sorted Loans (first 10):", sorted_loans[:10])
print("Index of Max Loan:", max_index)
print("Index of Min Loan:", min_index)
```

```python

random_loans = np.random.choice(loan_amount, 5)
print("Random Loan Samples:", random_loans)
```

```python
#  Reshape & Combine
matrix = np.column_stack((applicant_income[:5], loan_amount[:5]))
transposed_matrix = matrix.T
print("\nCombined Matrix (first 5 rows):\n", matrix)
print("Transposed Matrix:\n", transposed_matrix)
```
